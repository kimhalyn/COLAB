```
import matplotlib.pyplot as plt
import numpy as np
```

```
#기본차트
plt.title('plot1')
plt.plot([1, 3, 4, 8, 10, 17])
plt.show
```

```
#X축 설정
plt.title('plot2')
plt.plot([10, 20, 30, 40, 50, 60], [1, 3, 4, 8, 10, 17])
plt.show
```

```
#차트 스타일 - blue, squre, :점선, --좀 더 굵은 선
plt.title('plot3')
plt.plot([10, 20, 30, 40, 50, 60], [1, 3, 4, 8, 10, 17], 'bs:')
plt.grid(True)
plt.show()
```

```
#멀티 라인차트
vec = np.arange(0, 5, 0.2)

plt.title('plot4')
plt.plot(vec, vec, 'r--')
plt.plot(vec, vec**2, 'b--')
plt.plot(vec, vec**3, 'g--')
plt.grid(True)
plt.show()
```

```
#막대그래프
plt.title('plot5')
plt.bar([10, 20, 30, 40, 50, 60], [1, 3, 4, 8, 10, 17])
plt.xlabel('number')
plt.ylabel('count')
plt.show()
```

```
#히스토그램
data1 = np.random.rand(1000) #0 ~ 1 사이의 실수 1000개
data2 = np.random.randn(1000) #정규분포를 따르는 실수 1000개

plt.title('plot6')
plt.hist(data2, edgecolor='black')
plt.show()
```

```
#산점도
x = np.random.normal(0, 1, 1000)
y = np.random.normal(0, 1, 1000)

plt.title('Scatter')
plt.scatter(x, y)
plt.show()
```
