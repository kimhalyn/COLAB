```
#데이터프레임 인덱싱
#loc - 컬럼명 인덱싱 
df1 = df_exam.loc[:, ['math', 'science']]
df1

df2 = df_exam.loc[1:4, ['math', 'english']]
df2

#iloc - 인덱스 번호 인덱싱
df3 = df_exam.iloc[1:5, 2:4]
df3
```

```
#수학점수 60점 이상 데이터프레임 생성
df_math = df_exam[df_exam.math >= 60]
df_math
```

```
#수학점수 내림차순
df_math.sort_values(by='math', ascending=False)
```

```
#영어점수 60~70, 내림차순 정렬
df_english = df_exam[(df_exam.english >= 60) & (df_exam.english < 70)].sort_values(by='english', ascending = False)
df_english
```

```
#파생변수 - 기존의 변수에서 새로운 변수를 추가하는 것
df_exam['total'] = df_exam.math + df_exam.english + df_exam.science
df_exam

df_exam['mean'] = df_exam.total / 3
df_exam
```
